```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/B612_20210817_152122_517.jpg'  # Replace with the path to your image
image = Image.open(image_path)

# Display the original image
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
gray_image = image.convert('L')

# Convert grayscale image to a NumPy array
gray_array = np.array(gray_image)

# Optionally, convert NumPy array to a standard Python list
image_matrix = gray_array.tolist()
# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Convolution filter (3x3)
filter = [
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
]

# Function to apply the convolution filter to the image
def apply_convolution(image, filter):
    image_height = len(image)
    image_width = len(image[0])
    filter_height = len(filter)
    filter_width = len(filter[0])
    pad_height = filter_height // 2
    pad_width = filter_width // 2

    # Create a padded image
    padded_image = [[0] * (image_width + 2 * pad_width) for _ in range(image_height + 2 * pad_height)]
    for i in range(image_height):
        for j in range(image_width):
            padded_image[i + pad_height][j + pad_width] = image[i][j]

    # Create an output image
    output_image = [[0] * image_width for _ in range(image_height)]

    # Apply the convolution filter
    for i in range(pad_height, image_height + pad_height):
        for j in range(pad_width, image_width + pad_width):
            convolution_sum = 0
            for k in range(filter_height):
                for l in range(filter_width):
                    convolution_sum += padded_image[i - pad_height + k][j - pad_width + l] * filter[k][l]
            output_image[i - pad_height][j - pad_width] = convolution_sum

    return output_image

convolved_image = apply_convolution(gray_array, filter)

```

```python
# Function to display the images
def display_images(original, filter, convolved):
    # Create a figure to display images
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))

    # Original Image
    ax[0].imshow(original, cmap='gray')
    ax[0].set_title("Original Image")
    ax[0].axis('off')

    # Filter Image (for visualization, converting to an image)
    ax[1].imshow(filter, cmap='gray')
    ax[1].set_title("Convolution Filter")
    ax[1].axis('off')

    # Convolved Image
    ax[2].imshow(convolved, cmap='gray')
    ax[2].set_title("Convolved Image")
    ax[2].axis('off')

    plt.show()

# Display the original image, filter, and convolved image
display_images(gray_image, filter, convolved_image)
```
